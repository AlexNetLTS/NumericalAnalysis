```python
from Lagrange import lagrange
import numpy as np
from matplotlib_inline import backend_inline
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
# sns.set_palette('pastel')
backend_inline.set_matplotlib_formats('svg')

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
def f(x):
    return 1 / (1 + 25 * x**2)
```

```python
x = np.linspace(-1, 1, 10000)
```

```python
fig = plt.figure(figsize=(9, 6))
plt.xlim(-1, 1)
plt.ylim(0, 2)
plt.plot(x, f(x), label=r'原函数', lw=1.5)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('原函数与插值函数图像')
for i in [9, 19]:
    x1 = np.linspace(-1, 1, i + 1)
    plt.plot(x, [lagrange(a, x1, f(x1)) for a in x], lw=0.5, label=f'n={i + 1}')
plt.legend(loc='upper center', fontsize=10)
# plt.savefig(r'D:\PyCharm\PyPrograms\PyMath\data\img1.svg', dpi=1000)
```
